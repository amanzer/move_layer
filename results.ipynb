```python
import pickle
import matplotlib.pyplot as plt

plots= [["desktop", "multi_cores", "ais", "1000"],["desktop", "multi_cores", "ais", "5821"], ["desktop", "multi_cores", "stib", "100000"], ["desktop", "multi_cores", "stib", "497609"]]


for configs in plots:

    with open(f"{configs[0]}_results/{"multi_cores"}_results/{configs[2]}_{configs[3]}_multicores_fps_record.pkl", "rb") as f:
        fps_record_multi = pickle.load(f)


    with open(f"{configs[0]}_results/{"single_core"}_results/{configs[2]}_{configs[3]}_multicores_fps_record.pkl", "rb") as f:
        fps_record_single = pickle.load(f)

    plt.plot(fps_record_multi, label=f'Multi Cores FPS Record, avg FPS :{sum(fps_record_multi)/len(fps_record_multi)}')
    plt.plot(fps_record_single, label=f'Single Core FPS Record, avg FPS :{sum(fps_record_single)/len(fps_record_single)}')

    # Adding labels and title
    plt.xlabel('Frame')
    plt.ylabel('Seconds')
    plt.title(f"Configuration : {configs[0]}, {configs[2]}, {configs[3]}, Time Delta size = 30")

    # Adding legend
    plt.legend()
    plt.grid()

    # Display the plot
    plt.show()
    
```
