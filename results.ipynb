```python
import pickle
import matplotlib.pyplot as plt

config="desktop"
# config="laptop"
model="multi_cores"
# model="single_core"
dataset="ais"
dataset="stib"
# size="1000"
# size="5821"
# size="100000"
size="497609"

with open(f"{config}_results/{model}_results/{dataset}_{size}_multicores_fps_record.pkl", "rb") as f:
    fps_record = pickle.load(f)
with open(f"{config}_results/{model}_results/{dataset}_{size}_multicores_onf_record.pkl", "rb") as f:
    onf_record = pickle.load(f)
with open(f"{config}_results/{model}_results/{dataset}_{size}_multicores_matrix_cap_record.pkl", "rb") as f:
    matrix_cap_record = pickle.load(f)



# Plotting the records
plt.plot(fps_record, label='FPS Record')
plt.plot(onf_record, label='ONF Record')
plt.plot(matrix_cap_record, label='matrix_cap Record')

# Adding labels and title
plt.xlabel('Frame')
plt.ylabel('Seconds')
plt.title(f"Configuration : {size} {dataset}, {config}, {model}, Time Delta size = 30")

# Adding legend
plt.legend()

# Display the plot
plt.show()
```
